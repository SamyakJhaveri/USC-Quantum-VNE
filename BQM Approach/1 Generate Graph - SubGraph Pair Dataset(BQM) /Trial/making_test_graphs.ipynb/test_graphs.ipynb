```python
import networkx as nx  
```

```python
G1 = nx.Graph()
edges_1 = [(0, 1), (0, 3), (1, 2), (2, 3)]
G1.add_edges_from(edges_1)
nx.draw(G1, with_labels = True)
```

```python
G2 = nx.Graph()
edges_2 = [(0, 2), (0, 3), (1, 2), (1, 3)]
G2.add_edges_from(edges_2)
nx.draw(G2, with_labels = True)
```
